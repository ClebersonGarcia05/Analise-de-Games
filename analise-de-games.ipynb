```python
import requests
import pandas as pd

url = 'https://api.rawg.io/api' #url da API
endpoint = '/games' #dados de qual assunto quer puxar
""" endpoints disponíveis são:  
    /creator-roles
    /creators
    /developers
    /games
    /genres
    /platforms
    /publishers
    /stores
    /tags
    
    OBS: Caso o endpoint seja alterado o código deverá ser ajustado. 
"""
api_key = 'b3adab1939fc43a9be46de02a80894ae' #chave de autenticação para a API
params = {'key': api_key} #parametros que você quer passar para a API
response = requests.get(url + endpoint, params=params) #faz uma requisição para a url com o assunto e informa os parametros para autenticação
if response.status_code == 200: #verifica se a requisição foi realizada com sucesso
    data = response.json() #data recebe o json para análise
    games = data['results'] #games recebe os dados do campo results.
    
    #Extrair informações relevantes
    games_data = [] #criação de uma nova lista
    for game in games: #faz a iteração por volta dos campos de games
        game_info = { #cria um novo dicionário onde é colocado dados dos games
            'id': game['id'],
            'Nome': game['name'],
            'Data de Lançamento': game['released'],
            'Avaliação': game['rating'],
            'Plataformas': [platform['platform']['name'] for platform in game['platforms']],
            'Última Atualização': game['updated']
        }
        games_data.append(game_info) #Adiciona os dados de cada game na lista
    df = pd.DataFrame(games_data) #Transforma a lista em um DataFrame
    print(df) #Exibe o Dataframe
else: #Caso a requisição não obtiver sucesso ele imprime o erro e o código do erro.
    print(f'Erro na solicitação: {response.status_code}')
        
    
```
